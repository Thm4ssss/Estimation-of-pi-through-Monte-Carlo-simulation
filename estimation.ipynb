## Estimating the value of $\pi$ by performing a Monte-Carlo Simulation

The goal of this project is to estimate the value of $\pi$ by using a Monte-Carlo simulation. <br>
To do so, we consider a square of side $2R$, and we draw a circle of radius $R$ on its center. Thus, the area of the square is: $4R^2$. The area of the circle is $\pi R^2$.<br>

Now, let's consider a point in the square. The probability for the point to be also in the circle is equal to :
$$ P=\frac{\pi R^2}{4R^2}=\frac{\pi}{4} $$

We'll estimate this probability with a Monte Carlo simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameters of the simulation
N=1000
x=np.random.uniform(-1,1,N)
y=np.random.uniform(-1,1,N)

distance=np.square(x)+np.square(y)
number_in_circle=np.sum(distance<=1)
probability=number_in_circle/N
pi_value=4*probability

pi_value
```

By playing with the value of N, we can get closer to the actual value of $\pi$, but it costs more time to compute.
Let's do some vizualisation to conclude

```python
theta=np.linspace(0,2*np.pi,100)
rho=np.ones(100)

circle_x=np.multiply(rho,np.cos(theta))
circle_y=np.multiply(rho,np.sin(theta))

left_square_x=-1*np.ones(100)
left_square_y=np.linspace(-1,1,100)
top_square_x=np.linspace(-1,1,100)
top_square_y=np.ones(100)
bottom_square_x=np.linspace(-1,1,100)
bottom_square_y=-1*np.ones(100)
right_square_x=np.ones(100)
right_square_y=np.linspace(-1,1,100)


plt.figure(figsize=(8,8))
plt.plot(x,y,'x',color='k')
plt.plot(circle_x,circle_y,color='blue',label='Circle')
plt.plot(left_square_x,left_square_y,color='r')
plt.plot(top_square_x,top_square_y,color='r')
plt.plot(right_square_x,right_square_y,color='r')
plt.plot(bottom_square_x,bottom_square_y,color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Distribution of points')
plt.grid()
plt.legend()
plt.show()
```
